```python
# -----------------------------------------------------------------
# CELL 3: The PyTorch Neural Network
# -----------------------------------------------------------------
class QCAC_Surrogate(nn.Module):
    def __init__(self, num_buses, num_gens, num_cuts=5):
        super().__init__()
        self.num_buses = num_buses
        self.num_gens = num_gens
        self.num_cuts = num_cuts

        input_dim = num_buses * 2
        self.trunk = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU()
        )

        # HEAD 1: Learned Integer Cuts (A and b)
        # A: (num_cuts x num_gens), b: (num_cuts,)
        self.head_cuts = nn.Linear(64, (num_cuts * num_gens) + num_cuts)

        # HEAD 2: Direct Binary Prediction (For Phase 1 Pre-training)
        self.head_u = nn.Sequential(
            nn.Linear(64, num_gens),
            nn.Sigmoid()
        )

    def forward(self, pd_qd_tensor):
        feat = self.trunk(pd_qd_tensor)

        u_pred = self.head_u(feat)

        cuts_out = self.head_cuts(feat)
        A_cut = cuts_out[:, :self.num_cuts * self.num_gens].view(
            -1, self.num_cuts, self.num_gens)
        b_cut = cuts_out[:, self.num_cuts * self.num_gens:]

        return u_pred, A_cut, b_cut

num_cuts = 5
model = QCAC_Surrogate(num_buses, num_gens, num_cuts=num_cuts)
optimizer = optim.Adam(model.parameters(), lr=0.005)

print("Neural Network Architecture successfully initialized!")
print(model)
```

```python
# -----------------------------------------------------------------
# CELL 4: Phase 1 - Supervised Pre-Training
# -----------------------------------------------------------------
print("--- Phase 1: Supervised Pre-Training (Behavioral Cloning) ---")

epochs_phase1 = 600
criterion_bce = nn.BCELoss()

history_loss_u = []

model.train()
for epoch in range(epochs_phase1):
    optimizer.zero_grad()

    u_pred, A_cut, b_cut = model(X_sup_t)

    # BCE loss on the binary commitment decisions
    loss = criterion_bce(u_pred, Y_u_sup_t)

    loss.backward()
    optimizer.step()

    history_loss_u.append(loss.item())

    if (epoch + 1) % 40 == 0:
        print(f"Epoch {epoch+1:3d}/{epochs_phase1} | BCE Loss: {loss.item():.4f}")

print("Phase 1 Complete.")
```

```python

cvx_layer, cfg = build_cvxpy_ac_uc_layer_supervised(data, vr_init, vi_init, rho, xi_c, xij_c, x_ij_s, n_cuts=num_cuts)

# Pre-compute cost coefficient tensors once (mirrors the CVXPY layer objective)
cost_c2 = torch.tensor(cfg.cost_c2, dtype=torch.float32)
cost_c1 = torch.tensor(cfg.cost_c1, dtype=torch.float32)
cost_c0 = torch.tensor(cfg.cost_c0, dtype=torch.float32)

# Phase 2 uses a separate, much lower LR.
# Gradients through a CVXPY layer are dense and can be large;
# the Phase-1 lr=0.005 is far too aggressive here.
optimizer_p2 = optim.Adam(model.parameters(), lr=1e-4)

epochs_phase2 = 10
# lambda_integrality pushes u toward {0,1}.  Scale is O(1) (max 0.25 per gen),
# while gen_cost is O(100) per gen, so we normalise both by n_gen.
lambda_integrality = 10.0

model.train()
for epoch in range(epochs_phase2):
    epoch_loss = 0.0
    cost_avg   = 0.0
    int_avg    = 0.0
    valid_batches = 0
    t_start = time.time()

    for i in range(len(X_unsup_t)):
        x_in = X_unsup_t[i].unsqueeze(0)
        pd_t = x_in[0, :num_buses]   # (n_bus,) -- CvxpyLayer expects 1-D
        qd_t = x_in[0, num_buses:]

        optimizer_p2.zero_grad()

        _, A_cut, b_cut = model(x_in)
        A_cut_i = A_cut[0]   # (n_cuts, n_gens)
        b_cut_i = b_cut[0]   # (n_cuts,)

        try:
            u_out, pg_out, qg_out, vr_out, vi_out, c_ii_out, s_cut_out = cvx_layer(
                A_cut_i, b_cut_i, pd_t, qd_t,
                solver_args={"max_iters": 10000, "eps": 1e-3}
            )

            # True generation cost using per-unit cost coefficients.
            # Normalised by n_gen so it is comparable across grid sizes.
            gen_cost = (cost_c2 * pg_out**2 + cost_c1 * pg_out + cost_c0 * u_out).sum()
            gen_cost_norm = gen_cost / cfg.n_gen

            # Integrality regulariser: u*(1-u) is 0 at {0,1}, max 0.25 at 0.5.
            # NOTE: do NOT add sum(s_cut_out) here -- the CVXPY layer already
            # penalises cut-slack at CUT_PEN=100 internally, and the gradient of
            # s_cut w.r.t. (A_cut, b_cut) already flows back through the KKT
            # system.  Adding it to the training loss caused the loss explosion.
            integrality_pen = (u_out * (1.0 - u_out)).sum() / cfg.n_gen

            loss = gen_cost_norm + lambda_integrality * integrality_pen

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer_p2.step()

            epoch_loss += loss.item()
            cost_avg   += gen_cost_norm.item()
            int_avg    += integrality_pen.item()
            valid_batches += 1

        except Exception as e:
            pass

    elapsed  = time.time() - t_start
    n = max(valid_batches, 1)
    print(f"Epoch {epoch+1:2d}/{epochs_phase2} | "
          f"Loss: {epoch_loss/n:.4f} | "
          f"Cost/gen: {cost_avg/n:.2f} | "
          f"Integ/gen: {int_avg/n:.4f} | "
          f"Solved: {valid_batches}/{len(X_unsup_t)} | "
          f"{elapsed:.1f}s")

print("Phase 2 Complete.")
```

```python
# -----------------------------------------------------------------
# CELL 6: Phase 3 - Feasibility Restoration & Evaluation
# -----------------------------------------------------------------
print("--- Phase 3: Feasibility Restoration Projection ---")

def miqcqp_feasibility_restoration(data, u_frac, pd_dict, qd_dict):
    """
    Projects the continuous fraction (u_frac) to the closest physically 
    valid integer combination using an L1-norm objective.
    """
    from src.formulation import build_single_ac_uc_rectangular, _precompute_bus_connectivity
    
    # 1. Build the Exact, Non-Convex MINLP
    m_proj = build_single_ac_uc_rectangular(data)
    m_proj.setParam("OutputFlag", 0)
    m_proj.setParam("TimeLimit", 15.0)
    
    # 2. Update loads for this specific test case
    conn = _precompute_bus_connectivity(data)
    for b_id in data.buses:
        m_proj._pbal[b_id, 1].RHS = pd_dict.get(b_id, 0.0)
        m_proj._qbal[b_id, 1].RHS = qd_dict.get(b_id, 0.0)
        
    # 3. Add L1-Norm objective: min sum |u_int - u_frac|
    w = m_proj.addVars(list(data.gens.keys()), [1], lb=0, vtype=GRB.CONTINUOUS)
    for idx, g_id in enumerate(sorted(list(data.gens.keys()))):
        m_proj.addConstr(w[g_id, 1] >= m_proj._u[g_id, 1] - u_frac[idx])
        m_proj.addConstr(w[g_id, 1] >= u_frac[idx] - m_proj._u[g_id, 1])
        
    m_proj.setObjective(gp.quicksum(w[g, 1] for g in data.gens), GRB.MINIMIZE)
    m_proj.optimize()
    
    if m_proj.status in [GRB.OPTIMAL, GRB.SUBOPTIMAL]:
        u_final = [int(round(m_proj._u[g, 1].X)) for g in sorted(list(data.gens.keys()))]
        return u_final, "Optimal"
    else:
        return np.round(u_frac).astype(int).tolist(), "Heuristic Rounded"

# =================================================================
# PIPELINE INFERENCE TEST
# =================================================================
model.eval()

# Generate an unseen, heavily stressed test profile (1.18x load + noise)
test_scale = 1.18
test_pd = {}
test_qd = {}
test_input_list = []

for l_id, load in data.loads.items():
    b_id = load["load_bus"]
    noise_p = np.random.normal(1.0, 0.05)
    noise_q = np.random.normal(1.0, 0.05)
    test_pd[b_id] = load["pd"] * test_scale * noise_p
    test_qd[b_id] = load["qd"] * test_scale * noise_q

for b_id in data.buses:
    test_input_list.extend([test_pd.get(b_id, 0.0), test_qd.get(b_id, 0.0)])

x_test = torch.tensor(test_input_list, dtype=torch.float32).unsqueeze(0)
pd_test_t = x_test[0, :num_buses]   # (n_bus,) -- 1-D for CvxpyLayer
qd_test_t = x_test[0, num_buses:]

print("1. Running Neural Network + CVXPY Forward Pass...")
t0 = time.time()
with torch.no_grad():
    _, A_cut, b_cut = model(x_test)
    A_cut_i = A_cut[0]   # (n_cuts, n_gens)
    b_cut_i = b_cut[0]   # (n_cuts,)

    u_out, pg_out, qg_out, vr_out, vi_out, c_ii_out, s_cut_out = cvx_layer(
        A_cut_i, b_cut_i, pd_test_t, qd_test_t,
        solver_args={"max_iters": 5000, "eps": 1e-3}
    )
    u_frac_np = u_out.numpy()
t_forward = time.time() - t0

print(f"   -> Continuous Commitment (u_relax): {np.round(u_frac_np, 3)}")
print(f"   -> Forward Pass Time: {t_forward:.3f}s")

print("2. Running L1-Norm Feasibility Restoration...")
t0 = time.time()
u_final, status = miqcqp_feasibility_restoration(data, u_frac_np, test_pd, test_qd)
t_rest = time.time() - t0

print(f"   -> Final Integer Commitment: {u_final}")
print(f"   -> Restoration Status: {status}")
print(f"   -> Restoration Time: {t_rest:.3f}s")
print(f" Total End-to-End Inference Time: {t_forward + t_rest:.3f}s")
```
